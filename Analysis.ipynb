```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cleaned_df = pd.read_csv("./csv/cleaned.csv")

cleaned_df.head()
```

```python
corr_df = cleaned_df[['price', 'area', 'ppf']]
corr_df.head()
```

```python
# Correlation Matrix
corr_matrix = corr_df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', mask=mask, linewidths=.8, cmap='coolwarm')
plt.title('Correlation Matrix of Real Estate Offers')
plt.show()
```

```python
sns.boxplot(data=cleaned_df, y = cleaned_df['area'])
plt.title('Price Distribution')
print("Mean: ", cleaned_df['area'].mean())
print("Median: ", cleaned_df['area'].median())
print("1st quantile: ", cleaned_df['area'].quantile(0.25))
print("3rd quantile: ", cleaned_df['area'].quantile(0.75))
```

```python
pie_df = cleaned_df.groupby('region').count()

pie_df.head()

```

```python
plt.pie(pie_df['title'], labels=['HK Island', 'Kowloon', 'NT East', 'NT West'], autopct = '%.0f%%')
plt.title('Real Estate Offers District Distribution')
plt.show()
```

```python
group_df = cleaned_df.groupby('region')['ppf'].mean()
group_df
```

```python
group_df.plot.barh(xlabel = "Price per feet")
plt.title('Average Price Per Feet by Region')
plt.show()
```

```python
sns.regplot(data=cleaned_df, x='price', y='area')
plt.title('Price - Area correlation')
plt.xlabel('Price in Million HKD')
plt.show()

```
